```python
import matplotlib.pyplot as plt
import numpy as np

vector1 = np.random.rand(15)
vector2 = np.random.rand(15)

plt.plot(vector1, label='Vector 1')
plt.plot(vector2, label='Vector 2')

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Two Random Vectors')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

matrix = np.random.rand(4, 4)

row_labels = ['Row 1', 'Row 2', 'Row 3', 'Row 4']
col_labels = ['Col 1', 'Col 2', 'Col 3', 'Col 4']

plt.figure(figsize=(8, 6))
sns.heatmap(matrix, annot=True, cmap='viridis', fmt=".2f",
            xticklabels=col_labels, yticklabels=row_labels)

plt.title('4x4 Matrix Heatmap')
plt.show()
```

```python
A = np.random.randint(1, 10, (4, 4))
B = np.random.randint(1, 10, (4, 4))

add = A + B
sub = A - B
mult = np.dot(A, B)

print('Summation \n',add)
print('Subtraction \n',sub)
print('Multiplication \n',mult)

add_flat = add.flatten()
sub_flat = sub.flatten()
mult_flat = mult.flatten()

fig, axs = plt.subplots(1, 3, figsize=(20, 5))

axs[0].bar(range(len(add_flat)), add_flat, color='orange')
axs[0].set_title("Matrix A + B")
axs[0].set_ylim([0, 40])

axs[1].bar(range(len(sub_flat)), sub_flat, color='red')
axs[1].set_title("Matrix A - B")
axs[1].set_ylim([-20, 20])

axs[2].bar(range(len(mult_flat)), mult_flat, color='purple')
axs[2].set_title("Matrix A * B")
axs[2].set_ylim([0, 200])

plt.tight_layout()
plt.show()
```

Adds corresponding elements of two matrices. Maintains the same matrix size. Useful for combining data from different sources.
Subtracts elements of one matrix from another. Highlights differences between datasets. Often used in error analysis.
Not element-wise; follows matrix multiplication rules. Produces a new matrix with different dimensions. Used in transformations, linear equations, and AI models.
