```python
import numpy as np
```

**QUESTION**

Data: [5.2, 5.4, 5.1, 5.6, 5.3, 5.8]

Hypothesized mean: 5.5

Tasks:

* Calculate the t-statistic and p-value.

* Interpret and visualize the result at a 0.05 significance level.

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
data = np.array([5.2, 5.4, 5.1, 5.6, 5.3, 5.8])
hypothesized_mean = 5.5
significance_level = 0.05

t_stat, p_val = stats.ttest_1samp(data, hypothesized_mean)

print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_val:.4f}")

if p_val < significance_level:
    print("Null hypothesis rejected: The sample mean is significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference found.")

plt.figure(figsize=(5, 5))
plt.bar(1, np.mean(data), color="skyblue", label="Sample Mean")
plt.errorbar(1, np.mean(data), yerr=np.std(data), fmt='o', color="black", linewidth=2.5)
plt.axhline(y=hypothesized_mean, color="red", linestyle="--", linewidth=1.5, label="Hypothesized Mean")
plt.xlim(0.5,1.5)
plt.xlabel('Sample value')
plt.ylabel("Mean Value")
plt.title("Comparison of Sample Mean and Hypothesized Mean")
plt.legend()
plt.grid(True, linestyle="--")
plt.show()
```
